```python
from pyspark.sql import SparkSession
```

```python
spark=SparkSession.builder.appName("partion-ass").getOrCreate()
```

```python
df=spark.range(500000)
```

```python
print("initial partition",df.rdd.getNumPartitions())
```

```python
df_repartition=df.repartition(12)
```

```python
print("After partition",df_repartition.rdd.getNumPartitions())
```

```python
df_coalesce=df_repartition.coalesce(3)
```

```python
print("After Coalesce",df_coalesce.rdd.getNumPartitions())
```
