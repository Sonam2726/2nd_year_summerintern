```python
#4) Write a Python program to calculate the total, highest value, and lowest value from a tuple of integers without using the built-in sum(), max(), and min() functions.
t1= (12, 45, 7, 23, 56, 18)

total = 0
highest = t1[0]
lowest = t1[0]

for num in t1:
    total += num

    if num > highest:
        highest = num

    if num < lowest:
        lowest = num

print("Total =", total)
print("Highest =", highest)
print("Lowest =", lowest)
```

```python
#Write a Python program to filter a tuple . n = (3, 14, 7, 22, 9, 41, 18, 5), keep only values greater than 10
n = (3, 14, 7, 22, 9, 41, 18, 5)

result = ()

for num in n:
    if num > 10:
        result += (num,)
print(result)
```
