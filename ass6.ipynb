```python
d9={"name":["priya","reeve","jaya"],"Age":[20,21,20],"marks":[85,90,78]}
d8=pd.DataFrame(d9)
print("Original Dataframe")
print(d8)
# iteration using row
print("Method 1")
for index,row in d8.iterrows():
    print(index,row['name'],row['Age'])
print("Method2")
for row in d8.itertuples():
    print(row.name,row.Age)
print("selection row using condition")
print("Student marks greater than 80")
res=d8[d8['marks']>80]
print(res)

print("selection row using iloc")
print(d8.iloc[::-1])

print("Limited row selection using Given columns ")
print(d8.iloc[0:2,2])

print("Drop row based on condition")
print("removing row having marks less than 80")
new_df=d8[d8['marks']>=80]
print(new_df)

print("Create list from rows in DataFrame")
print(d8.values.tolist())
```
